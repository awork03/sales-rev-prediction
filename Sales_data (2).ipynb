```python
import pandas as pd
import numpy as np
data = {
    "name": ["Alice", "Ben", "Carlos", "Dana", "Elena", "Frank", "Grace", "Henry", "Isabella", "Jake"],
    "years_experience": [2, 5, 1, 4, 3, 6, 2, 7, 3, 5],
    "deals_closed": [12, 20, 8, 16, 14, 24, 10, 26, 13, 19],
    "total_sales": [120000, 210000, 75000, 190000, 160000, 260000, 110000, 280000, 140000, 200000],
    "region": ["East", "West", "South", "North", "West", "East", "South", "North", "East", "South"],
    "product_knowledge": [7.5, 9.0, 6.0, 8.5, 8.0, 9.5, 9.8, 7.8, 8.7, np.nan], # Added np.nan to make the list length 10
    "client_satisfaction": [4.2, 4.8, 3.5, 4.6, 4.3, 4.9, 4.0, 4.7, 4.1, 4.5],
    "calls_made": [250, 320, 180, 300, 270, 350, 240, 360, 260, 310],
    "revenue": [25000, 48000, 15000, 42000, 36000, 55000, 22000, 58000, 32000, 44000]
             }

df = pd.DataFrame(data)

df = pd.get_dummies(df, columns=['region'])

print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='deals_closed', y='total_sales', data=df)
plt.title('Relationship between Deals Closed and Total Sales')
plt.xlabel('Deals Closed')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.show()
```
